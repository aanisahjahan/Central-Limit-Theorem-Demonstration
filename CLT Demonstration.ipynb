# Import Necessary Libraries

```python
import scipy.stats as stats
import random
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Generating and Plotting Normal Distribution

```python
rnd_norm = []
np.random.seed(seed=10000)
for i in range(1000):
  rnd_norm.append(stats.norm.rvs(100,10))
ax = sns.histplot(rnd_norm, bins=30, stat='density')
```

# Sampling from the Normal Distribution

```python
num_sample = 1000
samplesize = 50
replacement = False

samples = []
for i in range(num_sample):
  samples.append([])
  if replacement == True:
    for j in range(samplesize):
      samples[i].append(random.choice(rnd_norm))
  else:
    samples[i].append(random.sample(rnd_norm,samplesize))

# Calculating and Plotting Sample Means (Normal Distribution)
sample_means = []
for k in samples:
  sample_means.append(np.mean(k))


ax = sns.histplot(sample_means, bins=20, stat='density')

mu, std = stats.norm.fit(sample_means)
xx = np.linspace(*ax.get_xlim(),100)
ax.plot(xx, stats.norm.pdf(xx, mu, std));
```
